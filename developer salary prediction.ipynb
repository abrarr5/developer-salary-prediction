```python
# Defining focused shortlist of technologies

languages = ['Python', 'JavaScript', 'TypeScript', 'Go', 'Rust']
clouds = ['AWS', 'GCP', 'Azure']
frameworks = ['React', 'Angular', 'Django', 'Flask']
devops_tools = ['Docker', 'Kubernetes', 'Terraform', 'GitHub Actions']
```

```python
# Binary indicators
for lang in languages:
    df_salary_filtered[lang] = df_salary_filtered['LanguageHaveWorkedWith'].str.contains(lang, na=False).astype(int)
for cloud in clouds:
    df_salary_filtered[cloud] = df_salary_filtered['PlatformHaveWorkedWith'].str.contains(cloud, na=False).astype(int)
for fw in frameworks:
    df_salary_filtered[fw] = df_salary_filtered['WebframeHaveWorkedWith'].str.contains(fw, na=False).astype(int)
for tool in devops_tools:
    df_salary_filtered[tool] = df_salary_filtered['ToolsTechHaveWorkedWith'].str.contains(tool, na=False).astype(int)
```

```python
# Dropping original string columns and interim count columns
cols_to_drop = [
    'LanguageHaveWorkedWith',
    'DatabaseHaveWorkedWith',
    'PlatformHaveWorkedWith',
    'WebframeHaveWorkedWith',
    'ToolsTechHaveWorkedWith',
    'Num_Languages',
    'Num_Databases',
    'Num_Platforms',
    'Num_Webframes',
    'Num_Tools',
    'Total_Skills'
]
df_salary_filtered = df_salary_filtered.drop(columns=cols_to_drop)
```

```python
print(df_salary_filtered.head())
print(df_salary_filtered.dtypes) 
```

```python
### Dataset Overview After Skill Encoding

# The dataset has been transformed so that all technical skills are represented as numeric or binary features. 
# Each selected language, cloud platform, framework, and DevOps tool is now a 0/1 indicator, capturing whether a respondent uses that technology. 
# Background and demographic information, such as age, education, experience, role, and organization size, remain as categorical or numeric fields. The salary variable (`ConvertedCompYearly`) is clean and ready for modeling. 
# This setup allows regression models to learn from interpretable numeric and boolean inputs while separating skill-related factors from other fixed characteristics.
```

```python

# Ordinal encode Age and Education

df_salary_filtered['Age'] = df_salary_filtered['Age'].astype(pd.CategoricalDtype(
    categories=['18-24 years old','25-34 years old','35-44','45-54','55-64','65+'], ordered=True)).cat.codes
df_salary_filtered['EdLevel'] = df_salary_filtered['EdLevel'].astype(pd.CategoricalDtype(
    categories=["Less than a bachelor's","Bachelor’s","Master’s","Professional","PhD"], ordered=True)).cat.codes
```

```python
# Simplified Employment

emp_map = {'Employed, full-time':'Full-time','Employed, part-time':'Part-time',
           'Independent contractor, freelancer, or self-employed':'Self-employed',
           'Not employed':'Other','Student':'Student'}
df_salary_filtered['Employment'] = df_salary_filtered['Employment'].map(emp_map).fillna('Other')
```

```python
# One-hot encode toggles

toggles = ['MainBranch', 'RemoteWork', 'AISelect', 'Industry']
toggles_existing = [col for col in toggles if col in df_salary_filtered.columns]

df_salary_filtered = pd.get_dummies(df_salary_filtered, columns=toggles_existing, drop_first=True)
```

```python
# Group DevType
dev_map = {'Developer, front-end':'Front-End','Developer, back-end':'Back-End','Developer, full-stack':'Full-Stack',
           'Data scientist or machine learning specialist':'Data','Data or business analyst':'Data',
           'Senior Executive (C-Suite, VP, etc.)':'Exec/Manager','Engineering manager':'Exec/Manager'}
df_salary_filtered['DevType_Group'] = df_salary_filtered['DevType'].replace(dev_map).fillna('Other')
df_salary_filtered = pd.get_dummies(df_salary_filtered, columns=['DevType_Group'], drop_first=True)
```

```python
# OrgSize ordinal
org_order = ['1-9','10-19','20-99','100-499','500-999','1,000-4,999','5,000-9,999','10,000+']
df_salary_filtered['OrgSize'] = df_salary_filtered['OrgSize'].astype(pd.CategoricalDtype(categories=org_order, ordered=True)).cat.codes
```

```python
# Drop high-cardinality free-text fields
drop_cols = ['LearnCode','OpSysProfessional use','MiscTechHaveWorkedWith','OpSysPersonal use','ProfessionalTech','Industry']
df_salary_filtered = df_salary_filtered.drop(columns=[c for c in drop_cols if c in df_salary_filtered.columns])
```

```python
# Convert to numeric, coerce errors to NaN
df_salary_filtered['YearsCode'] = pd.to_numeric(df_salary_filtered['YearsCode'], errors='coerce')
df_salary_filtered['YearsCodePro'] = pd.to_numeric(df_salary_filtered['YearsCodePro'], errors='coerce')
df_salary_filtered['WorkExp'] = pd.to_numeric(df_salary_filtered['WorkExp'], errors='coerce')
```

```python
# Drop any row with NaN in numeric features or target
df_salary_filtered = df_salary_filtered.dropna(subset=['YearsCode','YearsCodePro','WorkExp','ConvertedCompYearly'])
```

```python
# Compute correlations with the target
numeric_cols = df_salary_filtered.select_dtypes(include=['number','bool']).columns
corr_with_target = df_salary_filtered[numeric_cols].corr()['ConvertedCompYearly'].sort_values(ascending=False)
print(corr_with_target)
```

```python
# 1️⃣ Ensure target exists
target_col = 'ConvertedCompYearly'
if target_col not in df_salary_filtered.columns:
    raise ValueError(f"{target_col} not found in DataFrame!")

# 2️⃣ Ensure numeric/boolean features only
numeric_cols = df_salary_filtered.select_dtypes(include=['number','bool']).columns.tolist()

# 3️⃣ Compute correlations only on numeric/boolean columns
corr_with_target = df_salary_filtered[numeric_cols].corr()[target_col].sort_values(ascending=False)
print(corr_with_target)

# 4️⃣ Remove the target from correlation series for feature selection
feature_corr = corr_with_target.drop(target_col)

# 5️⃣ Drop weak predictors (absolute correlation < 0.01)
weak_features = feature_corr[feature_corr.abs() < 0.01].index.tolist()
df_salary_filtered = df_salary_filtered.drop(columns=weak_features)

# 6️⃣ Define X_final and y_final safely
X_final = df_salary_filtered.drop(columns=[target_col])
y_final = df_salary_filtered[target_col]

print("X_final shape:", X_final.shape)
print("y_final shape:", y_final.shape)
```

```python
df_salary_filtered = df_salary_filtered.drop(columns=['YearsCode','WorkExp'])
```

```python
X_final = df_salary_filtered.drop(columns=['ConvertedCompYearly'])  # features
y_final = df_salary_filtered['ConvertedCompYearly']                 # target
```

```python
# 1️⃣ Check which columns are still object type
str_cols = X_final.select_dtypes(include='object').columns.tolist()
print("String/object columns still in X_final:", str_cols)

# 2️⃣ One-hot encode all string columns
X_final_encoded = pd.get_dummies(X_final, columns=str_cols, drop_first=True)

# 3️⃣ Confirm all columns are numeric
print(X_final_encoded.dtypes.value_counts())
```

```python
# 4️⃣ Train/test split and fit LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

X_train, X_test, y_train, y_test = train_test_split(
    X_final_encoded, y_final, test_size=0.2, random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Metrics
print("Train R²:", r2_score(y_train, y_pred_train), "MAE:", mean_absolute_error(y_train, y_pred_train))
print("Test R²:", r2_score(y_test, y_pred_test), "MAE:", mean_absolute_error(y_test, y_pred_test))
```
